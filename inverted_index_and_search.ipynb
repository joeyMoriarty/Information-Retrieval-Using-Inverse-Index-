```python
porter=nltk.PorterStemmer()
df = [""]*34886
for i in range(0,len(d)):
    df[i] = d.iloc[i,0]
```

df contains the list of the documents now. But due to processing constraints I would be using a smaller segment of the same for creating the further inverse indexing and searching. A larger set or the whole of df can be used accordingly.

```python
df = df[0:100]
```

Creating the set of unique terms present in the document by first removing numbers and punctuations then tokenizing each word and then stemming it into lower case and base words then adding it to the set. 

```python
terms = set()
for doc in df:
    doc = re.sub(r'\d+', '', doc) #removing numbers
    doc = re.sub(r'[^\w\s]', '', doc) #removing punctuations
    for term in nltk.word_tokenize(doc):
        term = porter.stem(term)
        terms.add(term)
#terms
```

Now with the terms in hand we can heck each document for those terms and then create a inverse index matrix as the matrix contains the doc ids of all the termsn form of a dictionary.

```python
inv_index = {}

for term in terms:
    inv_index[term] = []

    for doc in df:
        if term in doc:
            inv_index[term].append(df.index(doc))
        else:
            continue
```

Creating a function which finds the intersection of two sets would be required when two queries are used for searching as I have used and as the boolean.

```python
def intersect(a, b):
    a_set = set(a)
    b_set = set(b)
 
    if (a_set & b_set):
        return(a_set & b_set)
    else:
        print("No documents with both queries try with individual queries ")
```

The main search function which takes in from the user if one input to be searched or two.
Then if one is to be searched it stems the input so that it matches with the keyword in our dictionary and hence we can find the list of docs from inverted index matrix.
The docs containing the queries are printed via the indices.

In case of two queries and requirement of both the queries we convert the lists into sets and then apply the intersect function defined above which gives us the doc ids which contain both the queries. ANd as in the previous case the relevant docs are printed.

```python
def search():
    gi = (input("Press 1 for single Query or 2 for two queries"))
    gi=int(gi)
    if gi == 1:
        q= input("enter your Query: ")
        q = porter.stem(q)
        if q in terms:
            v1 = inv_index[q]
            print("The relevant docs are as follows:")
            for i in v1:
                print(i,".  ",df[i],"\n",q,"\n")
        else:
            print("query not found :( ")
    elif gi == 2:
        q1= input("enter your Query1: ")
        q1= porter.stem(q1)
        q2= input("enter your Query2: ")
        q2= porter.stem(q2)
        if q1 in terms:
            v1 = np.array(inv_index[q1])
        else:
            print("Query one not present")
        if q2 in terms:
            v2 = np.array(inv_index[q2]) 
        else:
            print("Query two not present")
        v = intersect(v1,v2)
        if len(v)!=0:
            print("The doc ids for docs containing both queries are: \n", v, "\n")
            print("The relevant docs are as follows:")
            for j in v:
                print(j,".  ",df[j],"\n")
        
    else:
        
        print("enter valid input")
```

The above code can be scaled to more queries and a choice of different boolean operators can be added. 
Now run the code and search for query/ies based on the requirement.
